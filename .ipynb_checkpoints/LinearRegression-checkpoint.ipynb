```python
import numpy as np
Housing_data = np.array([list(map(float, l.split())) for l in open('housing.data')])
y = Housing_data[:,-1]
print ('Data points', len(y))
print (y.T)
```

```python
# Calcular la mitjana de les dades

import statistics as stats


Y_mean = stats.mean(y.T)

print(Y_mean)
```

```python
# Definim les operacions dot i invers
dot = np.dot
inv = np.linalg.inv
# Omplim la matriu de dades X en tots 1
X = np.ones((len(y),1))
# Calculem els parametres: theta = (X'*X)^-1*X'*y
theta = dot(dot(inv(dot(X.T, X)), X.T), y)
print(theta)
```

El valor de la mitjana i de theta son el mateix perque les dades de X només utilitzem el bias

```python
# MSE = (1/N)*sum((y-X*theta)^2) 

MSE = sum((y-dot(X, theta))**2) / len(y)
print(MSE)
```

```python
# Seperar tot el arxiu Housing_data entre 50% Train i 50% Test

PrimeresDades = len(Housing_data)/2

print('Primeres',PrimeresDades, 'dades')

DadesTotals = len(Housing_data)

print('Dades Totals:',DadesTotals)

Housing_data_train = Housing_data[0:253]

Housing_data_test = Housing_data[253:506]

y_train = Housing_data_train[:,-1]

print(len(y_train))

y_test = Housing_data_test[:,-1]

print(len(y_test))

# Omplim la matriu de dades_train X1 en tots 1
X = np.ones((253,1))
```

```python
# Definició de la matriu X amb totes les diferents dades + el bias i el calcul de la seva theta = (X'*X)^-1*X'*y

# Totes les variables + el bias

Allvar_train = np.hstack((X, Housing_data_train[:,0:14]))
Allvar_test = np.hstack((X, Housing_data_test[:,0:14]))

# El model (theta) amb totes les variables + el bias -> theta = (X'*X)^-1*X'*y

theta_allvar = dot(dot(inv(dot(Allvar_train.T, Allvar_train)), Allvar_train.T), y_train)

# Calcul de la MSE de totes les variables amb la seva theta corresponent (MSE_train i MSE_test)

# MSE = (1/N)*sum((y-X*theta)^2) 

# MSE_allvar

MSE_allvar_train = sum((y_train-dot(Allvar_train, theta_allvar))**2) / 253

MSE_allvar_test = sum((y_test-dot(Allvar_test, theta_allvar))**2) / 253

print('MSE_allvar_train:', MSE_allvar_train)
print('MSE_allvar_test:', MSE_allvar_test)
print()

# Calcul del coeficient de determinació (R²) de les variables test

# Variance (Var) = (1/N)*sum(y_test.mean-y_test)²

Var = np.var(y_test)

# FVU = MSE/Var

FVU_allvar_test = MSE_allvar_test/Var

# R² = 1 - FVU

Rsquare_allvar_test = 1 - FVU_allvar_test


# Mostrem els diferents coeficients de determinació de cada una de les diferents variables (MSE_test)

print('Rsquare_allvar_test:', Rsquare_allvar_test)
```

```python
# Definició de la matriu X amb totes les diferents dades menys la variable del apartat Q3 que feia que el sistema era pitjor + el bias i el calcul de la seva theta = (X'*X)^-1*X'*y

# Totes les variables menys la primera + el bias

Allvar_train_notvar1 = np.hstack((X, Housing_data_train[:,1:14]))
Allvar_test_notvar1 = np.hstack((X, Housing_data_test[:,1:14]))

# El model (theta) -> theta = (X'*X)^-1*X'*y

theta_allvar_notvar1 = dot(dot(inv(dot(Allvar_train_notvar1.T, Allvar_train_notvar1)), Allvar_train_notvar1.T), y_train)

# Calcul de la MSE amb la seva theta corresponent (MSE_train i MSE_test)

# MSE = (1/N)*sum((y-X*theta)^2) 

# MSE_allvar_notvar1

MSE_allvar_train_notvar1 = sum((y_train-dot(Allvar_train_notvar1, theta_allvar_notvar1))**2) / 253

MSE_allvar_test_notvar1 = sum((y_test-dot(Allvar_test_notvar1, theta_allvar_notvar1))**2) / 253

print('MSE_allvar_train_notvar1:', MSE_allvar_train_notvar1)
print('MSE_allvar_test_notvar1:', MSE_allvar_test_notvar1)
print()

# Calcul del coeficient de determinació (R²) de les variables test

# Variance (Var) = (1/N)*sum(y_test.mean-y_test)²

Var = np.var(y_test)

# FVU = MSE/Var

FVU_allvar_test_notvar1 = MSE_allvar_test_notvar1/Var

# R² = 1 - FVU

Rsquare_allvar_test_notvar1 = 1 - FVU_allvar_test_notvar1


# Mostrem els diferents coeficients de determinació de cada una de les diferents variables (MSE_test)

print('Rsquare_allvar_test_notvar1:', Rsquare_allvar_test_notvar1)
```
